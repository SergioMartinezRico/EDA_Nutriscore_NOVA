```python
import pandas as pd
import numpy as np
```

lectura el csv original

```python
df = pd.read_csv("Data/openfoodfacts_export.csv", sep="\t", low_memory=False)
df.info()
```

eliminar columnas que tenga todo nan o null

```python
df = df.dropna( how = "all")
```

```python
df.info()
```

Eliminar columnas que tengan menos del 30% de sus valores 

```python
#army = army.dropna(axis=1, thresh = int(0.7*len(army)))

df = df.dropna(axis=1, thresh = int(0.7*len(df)))
```

```python
df.info()
```

```python
df.to_csv("comidas_limpio_1.csv")
```

```python
# --- LIMPIEZA DE COLUMNAS IRRELEVANTES ---

# Copiamos el DataFrame original para trabajar con seguridad
df_clean = df.copy()

# Eliminamos columnas no útiles para el análisis del Nutri-Score
cols_to_drop = [
    # Identificadores y metadatos
    'code', 'url', 'creator', 'created_t', 'last_modified_t', 
    'update_at', 'scoring_version', 'unique_scans_n',

    # Imágenes y texto libre
    'image_url', 'image_small_url', 'image_front_url', 
    'packaging_text', 'ingredients_text', 'ingredients_text_en', 
    'labels_tags', 'traces_tags', 'additives_tags', 'brands_tags',

    # Columnas redundantes o duplicadas
    'countries_tags', 'countries_fr', 'countries_es', 
    'categories_tags', 'categories_fr', 'categories_es',

    # Tiendas y localización comercial
    'stores', 'purchase_places', 'emb_codes', 'packaging',

    # Otras poco informativas para este análisis
    'quantity', 'serving_size', 'serving_quantity', 'data_quality_tags'
]

# Eliminamos solo las que existan realmente (evita errores si faltan)
df_clean = df_clean.drop(columns=[c for c in cols_to_drop if c in df_clean.columns], errors='ignore')

# Mostramos cuántas columnas quedan
print(f"Columnas restantes tras la limpieza: {len(df_clean.columns)}")
print("Primeras columnas conservadas:")
print(df_clean.columns[:20])


df_clean.info()
```
